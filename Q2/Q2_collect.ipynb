```python
#2. Which combination of production budget, genre, runtime
#and release timing best explains financial and rating-based anime movie success in the last 26 years?

import os
import requests
import json
from datetime import datetime
from dotenv import load_dotenv
import time

load_dotenv("test.env")
api_key_TMDB = os.getenv("TMDB_API_KEY")


if not api_key_TMDB:
    print("API KEY not found check env file")
    exit()
else:
    print("KEY loaded")


```

```python
#Building our links for the different apis
tmdb_base = "https://api.themoviedb.org/3"
tmbd_endpoint = "/discover/movie"  #Get all result for "" fitting criteria
tmdb_url = tmdb_base + tmbd_endpoint

```

```python
API_KEY = ""


#tmdb_url = "https://api.themoviedb.org/3/discover/movie"

# Calculate 40 years ago
current_year = datetime.now().year
start_year = current_year - 5
start_date = f"{start_year}-01-01"

#Filter
parameters = {
    'api_key' : API_KEY,
    'with_genres' : 16,         #Genre Animation
    'with_original_language' : 'ja', #Language Japanese
    'primary_release_date.gte' : start_date,
    "vote_count.gte": 50,       #Skip over unimportant movies
    'sort_by' : 'primary_release_date.asc',
    'page' : 1 
} 

result = []
total_pages = 1

# Going through all pages
while parameters['page'] <= 1:
    tmdb_response=requests.get(tmdb_url, params=parameters)

    # Send request to api
    if tmdb_response.status_code == 200:
        tmdb_data = tmdb_response.json()
        total_pages = tmdb_data['total_pages']

        #For each movie per page
        for movie in tmdb_data['results']:

            # request link to get details about each movie, for runtime, revenue, budget
            details_url = f"https://api.themoviedb.org/3/movie/{movie['id']}"
            detail_params = {"api_key": API_KEY}

            # sending request
            try:
                detail_response = requests.get(details_url, params=detail_params)
                if detail_response.status_code == 200:
                    detail_data = detail_response.json()
            except:
                pass

            #add movie to result
            result.append({
                'title' : movie['title'],
                'release_date' : movie['release_date'],
                'rating' : movie['vote_average'],
                'id' : movie['id'],
                'runtime' : detail_data['runtime'],
                'revenue' : detail_data['revenue'],
                'budget' : detail_data['budget']
            })

            #Spam protection
            time.sleep(0.25)
            


        print(f"Fethed page {parameters['page']} of {total_pages}")
        
        #Going over to next page
        parameters['page'] += 1

        #Spam protection
        time.sleep(0.3)
    
    else:
        print(f"Error on page {parameters['page']}")
        break    

print(f"loaded {len(result)} movies.\n")

print("The first five movies on the list:")
for movie in result[:5]:
    print(movie)
```
