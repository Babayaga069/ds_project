```python
import requests

#Test queries for Steam Web API

#Number of current players for CS 2
steam_url = "https://api.steampowered.com/ISteamUserStats/GetNumberOfCurrentPlayers/v1/?appid=730"
steam_response = requests.get(steam_url)
print(steam_response.json()['response']['player_count'])

#Global Percentages per Achievements in CS 2
steam_url1 = "https://api.steampowered.com/ISteamUserStats/GetGlobalAchievementPercentagesForApp/v2/?gameid=730"
steam_response1= requests.get(steam_url1)
print(steam_response1.json())
```

```python
import requests
import json

#Test queries for Twitch API

#Current Top Games on Twitch
twitch_url = "https://api.twitch.tv/helix/games/top"
twitch_headers = {
    "Client-ID": "CLIENT_ID",
    "Authorization": "Bearer ACCESS_TOKEN"
}

twitch_response = requests.get(twitch_url, headers=twitch_headers).json()
print(json.dumps(twitch_response, indent=1))
```
