```python
import requests

#Test queries for Steam Web API

#Number of current players for CS 2
steam_url = "https://api.steampowered.com/ISteamUserStats/GetNumberOfCurrentPlayers/v1/?appid=730"
steam_response = requests.get(steam_url)
print(steam_response.json()['response']['player_count'])

steam_url1 = "https://api.steampowered.com/ISteamUserStats/GetGlobalAchievementPercentagesForApp/v2/?key=&gameid=620"
steam_response1= requests.get(steam_url1)
print(steam_response1.json())
```
